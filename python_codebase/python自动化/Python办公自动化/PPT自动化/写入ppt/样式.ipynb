# 设置文本框样式

```python
from pptx import Presentation
from pptx.util import Cm
from pptx.enum.text import MSO_VERTICAL_ANCHOR # 对齐方式
from pptx.dml.color import RGBColor
# 新建PPT
prs = Presentation()
# 添加一个幻灯片使用6号板式
slide = prs.slides.add_slide(prs.slide_layouts[6])

left= top  = heigth = Cm(3)
wdith = Cm(10)
# 添加一个文本框
textbox = slide.shapes.add_textbox(left,top, wdith, heigth)
tf = textbox.text_frame
tf.text = "asdfasdfasdf"

# 设置样式
# 距离下的
tf.margin_bottom = Cm(1)
# 距离左侧
tf.margin_left = Cm(1)
# 设置对齐方式
tf.vertical_anchor = MSO_VERTICAL_ANCHOR.BOTTOM # 设置底部对齐
tf.word_wrap = True # 设置自动换行

# 设置背景颜色，使用的是文本框对象
fill = textbox.fill
# 纯色填充
fill.solid() 
fill.fore_color.rgb = RGBColor(255,192,203)


# 设置文本框边框的颜色color,宽度width
# 使用文本框对象
line = textbox.line
# 设置颜色
line.color.rgb = RGBColor(100,100,100)
# 设置宽度
line.width = Cm(0.5)

prs.save("./pptx样式.pptx")




```

# 设置文本框中段落的样式 

```python
from pptx import Presentation
from pptx.util import Cm
from pptx.enum.text import MSO_VERTICAL_ANCHOR  ,PP_PARAGRAPH_ALIGNMENT# 对齐方式
from pptx.dml.color import RGBColor
from docx.shared import Pt 

# 新建PPT
prs = Presentation()
# 添加一个幻灯片使用6号板式
slide = prs.slides.add_slide(prs.slide_layouts[6])

left= top  = heigth = Cm(3)
wdith = Cm(10)
# 添加一个文本框
textbox = slide.shapes.add_textbox(left,top, wdith, heigth)
tf = textbox.text_frame
# 添加一个段落
p = tf.add_paragraph()
p.text = "阿萨德加客服规划师"

# 设置对其方式  这里居中对其
p.alignment =PP_PARAGRAPH_ALIGNMENT.CENTER

# 向段落中添加新的文字快
run = p.add_run()
run.text =  "阿萨德加客服号"
# 设置字体加粗
run.font.bold=True
# 设置倾斜
run.font.italic=True
# 设置下划线
run.font.underline = True
# 设置字体大小
run.font.size = Pt(20)
# 设置颜色
run.font.color.rgb = RGBColor(200,200,200)

# 添加段落 设置段前间距
p1 = tf.add_paragraph()
p1.text = "ddssss"
# 设置对其方式  这里居中对其
p1.alignment =PP_PARAGRAPH_ALIGNMENT.CENTER
# 设置行间距
p1.line_spacing = 2.0# 表示两倍行间距

# 设置段前间距
p1.space_after=Pt(10)
# 设置段后间距
p1.space_befor=Pt(20)



# 设置样式
# 距离下的
tf.margin_bottom = Cm(1)
# 距离左侧
tf.margin_left = Cm(1)
# 设置对齐方式
tf.vertical_anchor = MSO_VERTICAL_ANCHOR.BOTTOM # 设置底部对齐
tf.word_wrap = True # 设置自动换行

# 设置背景颜色，使用的是文本框对象
fill = textbox.fill
# 纯色填充
fill.solid() 
fill.fore_color.rgb = RGBColor(255,192,203)


# 设置文本框边框的颜色color,宽度width
# 使用文本框对象
line = textbox.line
# 设置颜色
line.color.rgb = RGBColor(100,100,100)
# 设置宽度
line.width = Cm(0.5)

prs.save("./pptx文本框段落样式.pptx")
```

# 设置段落字体的样式

```python
from pptx import Presentation
from pptx.util import Cm
from pptx.enum.text import MSO_VERTICAL_ANCHOR  ,PP_PARAGRAPH_ALIGNMENT# 对齐方式
from pptx.dml.color import RGBColor
from docx.shared import Pt 

# 新建PPT
prs = Presentation()
# 添加一个幻灯片使用6号板式
slide = prs.slides.add_slide(prs.slide_layouts[6])

left= top  = heigth = Cm(3)
wdith = Cm(10)
# 添加一个文本框
textbox = slide.shapes.add_textbox(left,top, wdith, heigth)
tf = textbox.text_frame
# 添加一个段落
p =  tf.add_paragraph()
p.text = "阿萨德加客服规划师"

p.alignment =PP_PARAGRAPH_ALIGNMENT.CENTER
# 设置段落的字体样式
p.font.name = "微软雅黑"
p.font.bold = True
p.font.size = Pt(30) 
p.font.color.rgb =RGBColor(255,192,203)

prs.save("设置字体样式.pptx")

```
