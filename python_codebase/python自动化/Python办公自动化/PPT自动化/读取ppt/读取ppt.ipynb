# 读取ppt

```python
from pptx import Presentation
pre = Presentation("./期末总结符.pptx")
slides = pre.slides # 获取所有幻灯片
for slide in slides:
    for shape in  slide.shapes: # 获取shape形状
        if shape.has_text_frame:# 判断这个形状有没有文字
            text_frame = shape.text_frame # 获取这个文本框
            for para in text_frame.paragraphs:# 获取段落
#                 print(text_frame.text) #获取内容
                print(para.text)
    print("__"*88)
```
