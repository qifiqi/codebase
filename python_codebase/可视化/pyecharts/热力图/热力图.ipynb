# 设置热力图

### 导入包，导入数据

```python
import random

from pyecharts import options as opts
from pyecharts.charts import HeatMap
from pyecharts.faker import Faker

value = [[i, j, random.randint(0, 50)] for i in range(24) for j in range(7)]
print(value)
c = (
    HeatMap(
    init_opts=opts.InitOpts(
        page_title='热力图',
    )
    )
    .add_xaxis(
        xaxis_data=Faker.clock)
    .add_yaxis(
        series_name='数据',
        yaxis_data=Faker.week,
        value=value,
        label_opts=opts.LabelOpts(#标签配置项
            is_show=True,
            position='inside'#标签的位置是否在图上显示值，并且值的位置
        )   
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title='热力图'),
        visualmap_opts=opts.VisualMapOpts()
    )
#     .render('热力图.html')
)
c.render_notebook()
```
