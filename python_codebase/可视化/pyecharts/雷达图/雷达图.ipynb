## 导包清洗数据，然后再用pyechters画出红小豆最近一年4个季度的价格平均值的大小

```python
import pandas as pd 
import numpy as np
from pyecharts.charts import Radar
from pyecharts import options as opts
```

## 清洗数据，取出2021年中数据为红小豆，并且处理好数据和价格

```python
data = pd.read_table('../数据/商品报价统计.txt',delimiter=',')
data_hxd=data[data['商品名字']=='红小豆'].copy()
data_hxd['价格'] = data_hxd['价格'].apply(lambda a :str(a).split('.')[0])
data_hxd['日期'] = data_hxd['日期'].astype(np.datetime64)
data_hxd['价格'] = data_hxd['价格'].astype(np.int64)
data_hxd.set_index('日期',inplace=True)
data_hxd = data_hxd['2021-12':'2021-01']
# 按照季度统计，默认每个季度的第一天
y = round(data_hxd.resample('QS').mean())
y = [[i for i in y['价格']]]

x = [0,0.5,1,1.5]
print(y)
```

```python
r=(
    Radar(
    init_opts=opts.InitOpts()#这个是设置初始画布
    )
    .add_schema(
        shape="circle",#这个是设置图形的样式，这里是圆形，如果没有设置这是看你的schema设置了几个值来绘制
        center=["50%", "50%"],#这个是设置距离中心点的位置，第一个是左右，第二个是上下这里是正中间
        schema=[#这里是四个角分别对应的属性以及
            opts.RadarIndicatorItem(name='第一季度',max_=20000),
            opts.RadarIndicatorItem(name='第二季度',max_=20000),
            opts.RadarIndicatorItem(name='第三季度',max_=20000),
            opts.RadarIndicatorItem(name='第四季度',max_=20000),
        ],
        splitarea_opt=opts.SplitAreaOpts(
            is_show=True,
            areastyle_opts=opts.AreaStyleOpts(opacity=0.5),#图形透明度。支持从 0 到 1 的数字，为 0 时不绘制该图形。
        ),
        textstyle_opts=opts.TextStyleOpts(
            # 文本配置项
            color='#0000CD',
        ),
    )
    .add(
        series_name='红小豆',#这个是设置这个名字
        data=y,#数据
        linestyle_opts=opts.LineStyleOpts(#设置线的样式
            color='#00FFFF',
        ),
        
    )
    .set_series_opts(#系列配置项
        label_opts=opts.LabelOpts(
            is_show=True,#标签
        ),
    )
    .set_global_opts(#全局
        #标题
        title_opts=opts.TitleOpts(
            title='红小豆一年变化雷达图',),
#         图裂配置
        legend_opts=opts.LegendOpts(),
    )
    .render('红小豆一年变化雷达图.html')
)
# r.render_notebook()
```

```python
from pyecharts import options as opts
from pyecharts.charts import Radar

data = [{"value": [4, -4, 2, 3, 0, 1], "name": "预算分配"}]
c_schema = [
    {"name": "销售", "max": 4, "min": -4},
    {"name": "管理", "max": 4, "min": -4},
    {"name": "技术", "max": 4, "min": -4},
    {"name": "客服", "max": 4, "min": -4},
    {"name": "研发", "max": 4, "min": -4},
    {"name": "市场", "max": 4, "min": -4},
]
c = (
    Radar()
    .set_colors(["#4587E7"])
    .add_schema(
        schema=c_schema,
        shape="circle",
        center=["50%", "50%"],
        radius="80%",
        angleaxis_opts=opts.AngleAxisOpts(
            min_=0,
            max_=360,
            is_clockwise=False,
            interval=5,
            axistick_opts=opts.AxisTickOpts(is_show=False),
            axislabel_opts=opts.LabelOpts(is_show=False),
            axisline_opts=opts.AxisLineOpts(is_show=False),
            splitline_opts=opts.SplitLineOpts(is_show=False),
        ),
        radiusaxis_opts=opts.RadiusAxisOpts(
            min_=-4,
            max_=4,
            interval=2,
            splitarea_opts=opts.SplitAreaOpts(
                is_show=True, areastyle_opts=opts.AreaStyleOpts(opacity=1)
            ),
        ),
        polar_opts=opts.PolarOpts(),
        splitarea_opt=opts.SplitAreaOpts(is_show=False),
        splitline_opt=opts.SplitLineOpts(is_show=False),
    )
    .add(
        series_name="预算",
        data=data,
        areastyle_opts=opts.AreaStyleOpts(opacity=0.1),
        linestyle_opts=opts.LineStyleOpts(width=1),
    )
#     .render("radar_angle_radius_axis.html")
)
c.render_notebook()
```
