# 地图中国——治愈数地图

#### 导入模块
#### 并清洗数据，这一次要把治愈和死亡人数，以及累计确诊写入，设置图例为单选模式

```python
import pandas as pd
from pyecharts.charts import Map
from pyecharts import options as opts
```

### 处理数据 

```python
data = pd.read_csv('./疫情数据全国.csv')
# data.set_index('城市',inplace=True)
data_q = data[['城市','累计确诊']].copy()
data_z = data[['城市','治愈']].copy()
data_kill = data[['城市','死亡']].copy()

data_q = [[i[0],i[1]] for i in data_q.values]
data_z = [[i[0],i[1]] for i in data_z.values]
data_kill = [[i[0],i[1]] for i in data_kill.values]
```

```python
bm=(
    Map(init_opts=opts.InitOpts(
        page_title='中国地图疫情数据',
        )
    )
    .add(
        is_map_symbol_show=False,
        series_name='累计确诊',
        data_pair=data_q,
        is_selected=True,  # 是否选中图例
    )
    .add(
        is_map_symbol_show=False,
        series_name='治愈',
        data_pair=data_z,
    )
    .add(
        is_map_symbol_show=False,
        series_name='死亡',
        data_pair=data_kill,
    
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title='疫情数据',
            title_textstyle_opts=opts.TextStyleOpts(
                color='#FF00FF',
                font_family='monospace',
                font_style='normal')
        ),
        legend_opts=opts.LegendOpts(
            type_='plain',
            selected_mode='single',
            textstyle_opts=opts.TextStyleOpts(
                color='#00CED1',
            ),
        ),
        visualmap_opts=opts.VisualMapOpts(
            is_show=True,
            min_=20,
            max_=2000,
            range_color=['#4682B4','#00BFFF','#20B2AA','#7CFC00','#FF7F50'],
        ),   
    )
    
)
bm.render_notebook()
```
