# 词云图的画法

### 第一步导包及相关模块

```python
from pyecharts.charts import WordCloud
import pandas as pd
from pyecharts import options as opts

data = pd.read_json('../地图/疫情数据全国.json')

c = (
    WordCloud()
    .add(
        series_name='',#这个是图列的名字，这个我并没有设置
        
        data_pair=data.values,#当传入的是json格式的时候可以直接使用，但是要对应起来
        
        shape='cardioid',
#         shape='diamond',#这些都是形状，词云图的性状
#         shape='triangle-forward',
        
        rotate_step='120', #旋转单词角度
        
        word_gap= 50,#这个是设置每一个词之间的间隙的
        
        word_size_range=[30,80],#单词字体大小范围，并设置了最大最小值，是一个范围
        
        textstyle_opts=opts.TextStyleOpts(font_family='cursive'),#词云图文字的配置
        
    )
    
    .set_global_opts(#设置全局配置项的地方
        
        title_opts=opts.TitleOpts(#这个是标题的配置项
        
        title='疫情词云图',#这个用来设置标题
        
        title_textstyle_opts=opts.TextStyleOpts(font_size=30)#这个是设置标题的文字大小
    )
    )
#     .render('疫情词云图.html')#想要生成的html文件名字-这个是用来生成html文件的一个方法，也可以用来显示
)
c.render_notebook()
```
